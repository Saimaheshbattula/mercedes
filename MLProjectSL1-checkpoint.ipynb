## Import the Libraries required

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Read the data sets as pandas data frames 

```python
train_data=pd.read_csv('train.csv')
test_data=pd.read_csv('test.csv')
```

## Shape of data sets

```python
print(train_data.shape) # Find Number of rows and columns
print(train_data.columns)

print(test_data.shape) # Find Number of rows and columns
print(test_data.columns)

train_data.head() # Show first 5 records
```

```python
# variance
train_data.var()
```

## Removing the featurres with variance 0 from the train data set

```python
variance_with_zero = train_data.var()[train_data.var()==0].index.values
variance_with_zero
```

```python
# Drop zero variance variables

train_data = train_data.drop(variance_with_zero, axis=1)
```

```python
# As we don't need ID column

train_data = train_data.drop(['ID'], axis=1)
```

```python
train_data.shape
```

```python
train_data.isnull().sum().values
```

```python
train_data.head()
```

## Label Encoding

```python

from sklearn.preprocessing import LabelEncoder
for col in train_data.columns:
    # Check if column contains string values
    if train_data[col].dtype == 'object':
        # Create a LabelEncoder instance
        label_encoder = LabelEncoder()
        
        # Fit and transform the column
        train_data[col] = label_encoder.fit_transform(train_data[col].astype(str))
```

```python
train_data.head()
```

## Dimensionality Reduction (Train Data Set)

```python
from sklearn.model_selection import train_test_split
X = train_data.drop('y', axis=1)
y = train_data.y
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.3,random_state=42)
```

## Accuracy Metrics

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
pca_xtest = PCA(n_components=0.95)
pca_xtest.fit(xtest)
pca_xtest_data_transformed = pca_xtest.transform(xtest)
print(pca_xtest_transformed.shape)
y_pred_eval=model.predict(pca_xtest_data_transformed)
y_pred_eval
```

```python
# Calculate mean squared error (MSE)
mse = mean_squared_error(ytest, y_pred_eval)
print("Mean Squared Error:", mse)

# Calculate mean absolute error (MAE)
mae = mean_absolute_error(ytest, y_pred_eval)
print("Mean Absolute Error:", mae)

# Calculate R-squared
r2 = r2_score(ytest, y_pred_eval)
print("R-squared:", r2)
```
